```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
# linspace 100,that is values from 0-10 make 100 values
y=np.sin(x)
z=np.cos(x)
```

**Plotting the data**

```python
#sin wave
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# cosine wave
plt.plot(x,z)

plt.show()
```

**PLotting a parabola(graphical representation of a quadratic equation)**

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

**Bar Plot**

```python
fig=plt.figure()
# ax means axis
# ax=fig.add_axes([0,0,1,1]) we are plotting a rectangle
ax=fig.add_axes([0,0,1,1])
languages=['English','Swahili','Spanish','Latin','Germany']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUUMBER OF PEOPLE')
plt.show()
```

**Pie Chart**

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','Swahili','Spanish','Latin','Germany']
people=[100,50,150,40,70]
# autopct-to tell the plot how many decimal points we need
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

**Scatter Plot**

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

**3D Scatter Plot**

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20* np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
